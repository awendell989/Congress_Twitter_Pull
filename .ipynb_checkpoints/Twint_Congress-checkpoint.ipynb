```python
import pandas as pd
import sys
import twint
import datetime
import nest_asyncio
import os
from dateutil.relativedelta import relativedelta
nest_asyncio.apply()
import psycopg2
import sqlalchemy 
```

```python
class Twint_Loop_to_DB:
    def __ininit___(Twint_Search,search_start,search_end):
        self.start = search_start
        self.end = search_end
        self.search = Twint_Search
    def Twint_Month_Create(self):
        today = datetime.datetime.now().date()
        First_of_Current_month = today.replace(day = 1)
        months = pd.date_range(self.start,self.end, freq='MS').tolist()
        for x in months:
            directory_month = str(x.strftime("%Y-%b"))
            start = x
            start = x.date()
            end = start + relativedelta(months=+1)
            end = end
            directory = os.path.join(os.getcwd(),self.search)
            file = os.path.join(directory,directory_month+self.search+'.csv')
            if (start == First_of_Current_month and os.path.exists(file)):
                    os.remove(file)
            if  os.path.exists(file):
                print(file + ' exists!')
            else:
                if os.path.exists(directory) == False:
                    os.makedirs(directory)
                c = twint.Config()
                c.Username = self.search
                #c.Search = self.search
                fields = ["tweet","id","date","time","username","replies_count","retweets_count","likes_count"]
                c.Custom["tweet"] = fields
                #Timeline
                c.Since = str(start)
                c.Until = str(end)

                #Output
                c.Store_csv= True
                c.Output = file

                # Run
                twint.run.Search(c)
    def Twint_Month_Merge(self):
        #Directory
        directory = os.path.join(os.getcwd(),self.search)
        #Merge Data 
        Merge_Start = 'Full' #Do Not Change
        Merge_File_Name = Merge_Start + self.search + '.csv'
        Merge_File_Full = os.path.join(directory,Merge_File_Name)
        #Get the Files
        all_filenames = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
        all_filenames = [x for x in all_filenames if not x.startswith(Merge_Start)]
        #Combine Them All 
        combined_csv = pd.concat([pd.read_csv(os.path.join(directory,f)) for f in all_filenames ])
        #Fix the Data
        combined_csv = combined_csv[['id', 'conversation_id', 'created_at', 'date', 'time', 'timezone',
       'user_id', 'username', 'name', 'place', 'tweet', 'mentions', 'urls',
       'photos', 'replies_count', 'retweets_count', 'likes_count', 'hashtags',
       'cashtags', 'link', 'retweet', 'quote_url', 'video', 'near', 'geo',
       'source', 'user_rt_id', 'user_rt', 'retweet_id', 'reply_to',
       'retweet_date']]
        combined_csv.drop_duplicates(subset = 'id',inplace = True,keep = 'last')
        combined_csv['tdatetime'] = pd.to_datetime(combined_csv['date'] + ' ' + combined_csv['time'])
        combined_csv['keyword'] = self.search
        #Send Out
        combined_csv.to_csv(Merge_File_Full, index=False, encoding='utf-8-sig')
        #combined_csv.date = pd.to_datetime(combined_csv.date)
        self.resuting_merge = combined_csv
    def Send_to_DB(self):
        import sqlalchemy 
        engine = sqlalchemy.create_engine('postgresql://averyw:Ib4MR46ML49C1c1J9Ruhq6GB9@tascdata.ccgxdeimxmld.us-east-2.rds.amazonaws.com:5432/Twitter')
        self.resuting_merge.to_sql(self.search, engine,index = False,if_exists = 'replace',dtype ={'id':sqlalchemy.types.BIGINT,
'conversation_id':sqlalchemy.types.VARCHAR,
'created_at':sqlalchemy.types.VARCHAR,
'date':sqlalchemy.types.DATE,
'time':sqlalchemy.types.TIME,
'timezone':sqlalchemy.types.VARCHAR,
'user_id':sqlalchemy.types.VARCHAR,
'username':sqlalchemy.types.VARCHAR,
'name':sqlalchemy.types.VARCHAR,
'place':sqlalchemy.types.VARCHAR,
'tweet':sqlalchemy.types.VARCHAR,
'mentions':sqlalchemy.types.VARCHAR,
'urls':sqlalchemy.types.VARCHAR,
'photos':sqlalchemy.types.VARCHAR,
'replies_count':sqlalchemy.types.INTEGER,
'retweets_count':sqlalchemy.types.INTEGER,
'likes_count':sqlalchemy.types.INTEGER,
'hashtags':sqlalchemy.types.VARCHAR,
'cashtags':sqlalchemy.types.VARCHAR,
'link':sqlalchemy.types.VARCHAR,
'retweet':sqlalchemy.types.VARCHAR,
'quote_url':sqlalchemy.types.VARCHAR,
'video':sqlalchemy.types.VARCHAR,
'near':sqlalchemy.types.VARCHAR,
'geo':sqlalchemy.types.VARCHAR,
'source':sqlalchemy.types.VARCHAR,
'user_rt_id':sqlalchemy.types.VARCHAR,
'user_rt':sqlalchemy.types.VARCHAR,
'retweet_id':sqlalchemy.types.VARCHAR,
'reply_to':sqlalchemy.types.VARCHAR,
'retweet_date':sqlalchemy.types.VARCHAR,
'tdatetime':sqlalchemy.types.DateTime,
'keyword':sqlalchemy.types.VARCHAR} )
        print(self.search + ' sent to DB!')
```

# Loop

```python
Words_to_Scrape = pd.read_csv(os.path.join(os.getcwd(),'Words_to_Scrape.csv'))
Words_to_Scrape['End'].replace({'Present':datetime.datetime.now().date()},inplace = True)
Words_to_Scrape['End'] = pd.to_datetime(Words_to_Scrape['End'])
Words_to_Scrape['Start'] = pd.to_datetime(Words_to_Scrape['Start'])
```

```python
for i in Words_to_Scrape.index:
    Twint_Run = Twint_Loop_to_DB()
    Twint_Run.search = Words_to_Scrape.loc[i,'Word']
    Twint_Run.start = Words_to_Scrape.loc[i,'Start']
    Twint_Run.end = Words_to_Scrape.loc[i,'End']
    if Twint_Run.end == 'Present':
            Twint_Run.end = datetime.datetime.now().date()
    Twint_Run.Twint_Month_Create()
    Twint_Run.Twint_Month_Merge()
    Twint_Run.Send_to_DB()
```
